# Temperature, Pressure, and Chemistry

A TauREx 3 model is assembled from interchangeable physical components. Before simulating spectra, the atmosphere itself must be defined by addressing:
- The temperature profile.
- The pressure range over which the atmosphere extends.
- The gases present and their mixing ratios.

These core components define a planet's atmospheric properties and form the foundation for transmission and emission forward models.

## 2. Inspecting Pressure and Temperature profiles
    
Verifying the number of layers, pressure boundaries, and temperature values ensures the atmospheric structure is configured correctly. The properties of the generated temperature and pressure objects provide a summary of the atmospheric grid.

```python
iso_t, press

print(f'Layers: {press.nLayers}')
print(f'Pressure range: {press.profile.min():.3e} Pa to {press.profile.max():.3e} Pa')
print(f'Isothermal temperature: {iso_t.isoTemperature} K')
```

```python
pressure_bar = press.profile / 1e5

print('First five pressure levels in bar:')
print(pressure_bar[:5])
print('Last five pressure levels in bar:')
print(pressure_bar[-5:])
```

## 3. Chemistry and Gas Mixtures
    
Chemistry components define the composition of the atmosphere. TauREx 3 distinguishes between two types of gases:
- **Active gases**: Gases that contribute explicit opacities (absorption features) to the spectrum.
- **Inactive (Fill) gases**: Gases that establish the background atmosphere (e.g., H2, He) and contribute primarily to Collision-Induced Absorption (CIA) or Rayleigh scattering.

These lists dictate which molecular features will appear in the resulting spectrum.

```python
print(f'Active gases: {chemistry.activeGases}')
print(f'Inactive gases: {chemistry.inactiveGases}')

for gas_name, mix_ratio in zip(chemistry.gases, chemistry.mixProfile[:, 0]):
    print(f'{gas_name}: {mix_ratio:.3e}')
```

```python
layer0_total = chemistry.mixProfile[:, 0].sum()
print(f'Total mixing ratio in the first layer: {layer0_total:.6f}')
```
