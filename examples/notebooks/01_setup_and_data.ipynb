## Downloading Opacity files

TauREx model physics assumes the external data exists somewhere that feeds it. 
To run through the other examples, you can use the `ensure_opacity_data` helper function (provided in our examples directory), which will download a small example dataset if it doesn't already exist and configure the TauREx caches to use it. Running this code is safe: existing files are reused rather than downloaded again.

```python
ensure_opacity_data(download=True)

xsec_files = sorted(path.name for path in XSEC_DIR.glob('*.h5'))
cia_files = sorted(path.name for path in CIA_DIR.glob('*.cia'))

print('Opacity data are available and the TauREx caches are configured.')
print(f'Found {len(xsec_files)} cross-section files: {xsec_files}')
print(f'Found {len(cia_files)} CIA files: {cia_files}')
```
