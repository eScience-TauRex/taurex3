# Setup and Opacity Data

This notebook is the entry point for the example series. Before we ask TauREx to calculate anything, we need to answer three very practical questions:

- are we importing the local TauREx source tree we think we are?
- where will downloaded example data live?
- can TauREx actually see the opacity files it needs?

Those checks are not glamorous, but they prevent most of the confusing failures that appear later when model construction is otherwise correct.

```python
from _shared import CIA_DIR, PROJECT_ROOT, SRC_DIR, TMP_DIR, XSEC_DIR, ensure_opacity_data

print(f'Project root: {PROJECT_ROOT}')
print(f'Source directory: {SRC_DIR}')
print(f'Temporary data directory: {TMP_DIR}')
print(f'Cross-sections directory: {XSEC_DIR}')
print(f'CIA directory: {CIA_DIR}')

print('\nIf these paths look sensible, the notebook is pointing at the repository you expect.')
```

The paths printed below are worth reading rather than scrolling past. If the source or temporary-data directories are wrong here, every later notebook will be reasoning from the wrong environment.

TauREx forward models are only as useful as the opacity data behind them. In the full walkthrough this setup happens as part of a longer story; here we keep it focused but still make the purpose explicit.

The next cell downloads the small example dataset if needed and configures the TauREx caches to use it. Re-running it is safe: existing files are reused rather than downloaded again.

```python
ensure_opacity_data(download=True)

xsec_files = sorted(path.name for path in XSEC_DIR.glob('*.h5'))
cia_files = sorted(path.name for path in CIA_DIR.glob('*.cia'))

print('Opacity data are available and the TauREx caches are configured.')
print(f'Found {len(xsec_files)} cross-section files: {xsec_files}')
print(f'Found {len(cia_files)} CIA files: {cia_files}')
```
