# Downloading Test Data

The examples in this documentation require external data files:
- **Molecular cross-sections (opacities)** — wavelength-dependent absorption data for each molecule (e.g., H2O, CH4).
- **Collision-Induced Absorption (CIA) files** — pressure-broadened continuum opacities for molecular pairs (e.g., H2-H2, H2-He).

TauREx supports several cross-section formats: [ExoMolOP](https://www.aanda.org/articles/aa/full_html/2021/02/aa38350-20/aa38350-20.html) HDF5, legacy TauREx 2 pickle, and [ExoTransmit](https://github.com/elizakempton/Exo_Transmit). For CIA, both HITRAN `.cia` and legacy `.db` formats are supported. Additional formats can be added via plugins.

Important: TauREx provides the software to read and analyse these datasets, but the datasets themselves are third-party products. In particular, the opacity files downloaded below come from [ExoMol](https://www.exomol.com) and the CIA files come from [HITRAN](https://hitran.org/); they are not created, owned, or warranted by the TauREx team, and responsibility for the data content remains with the original providers.

The cells below download example HDF5 cross-sections from ExoMol and CIA files from HITRAN, then configure the TauREx caches to locate them.

```python
from _shared import CIA_DIR, PROJECT_ROOT, SRC_DIR, TMP_DIR, XSEC_DIR, ensure_opacity_data

print(f'Project root: {PROJECT_ROOT}')
print(f'Source directory: {SRC_DIR}')
print(f'Temporary data directory: {TMP_DIR}')
print(f'Cross-sections directory: {XSEC_DIR}')
print(f'CIA directory: {CIA_DIR}')
```

## Downloading Opacity Files

The `ensure_opacity_data` helper downloads a small example dataset if it does not already exist and configures the TauREx caches to use it. These opacity and CIA files are fetched from external providers for demonstration purposes; they are not TauREx data products, and TauREx only provides the software used to read and analyse them. Existing files are reused rather than downloaded again.

```python
ensure_opacity_data(download=True)

xsec_files = sorted(path.name for path in XSEC_DIR.glob('*.h5'))
cia_files = sorted(path.name for path in CIA_DIR.glob('*.cia'))

print('Opacity data are available and the TauREx caches are configured.')
print(f'Found {len(xsec_files)} cross-section files: {xsec_files}')
print(f'Found {len(cia_files)} CIA files: {cia_files}')
```

## Inspecting a Cross-Section

Individual cross-section objects can be retrieved from the cache by molecule name. The `opacity(temperature, pressure)` method returns the absorption cross-section (in cm$^2$/molecule) on the native wavenumber grid for any temperature and pressure.

```python
import matplotlib.pyplot as plt
import numpy as np

h2o_xsec = OpacityCache()['H2O']
wlgrid = 10000 / h2o_xsec.wavenumberGrid

plt.figure(figsize=(7, 4))
plt.plot(wlgrid, h2o_xsec.opacity(2000, 1e1), lw=0.3)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Wavelength (um)')
plt.ylabel('Cross-section (cm$^2$/molecule)')
plt.title('H$_2$O opacity at T = 2000 K, P = 10 Pa')
plt.grid(alpha=0.2)
```
